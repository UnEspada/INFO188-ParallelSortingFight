```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV sin encabezado
tiempos_df = pd.read_csv('../pruebas/tiempos_vs_n_cuda.csv', header=None)

# Asignar nombres de columnas manualmente
tiempos_df.columns = ['n', 'tiempo']

# Verificar las primeras filas para asegurarnos que las columnas se hayan asignado correctamente
print(tiempos_df.head(10))

# Gráfico de Tiempo vs n
plt.figure(figsize=(10, 6))
plt.plot(tiempos_df['n'], tiempos_df['tiempo'], marker='o', linestyle='-', color='b', label='Tiempo de ejecución')
plt.title('Tiempo vs Tamaño de Entrada (n)', fontsize=14)
plt.xlabel('Tamaño de Entrada (n)', fontsize=12)
plt.xscale('log')  # Escala logarítmica en el eje X
plt.yscale('log')
plt.xticks([100, 1000, 10000, 100000, 1000000, 10000000], ['100', '1k', '10k', '100k', '1M', '10M'])
plt.yticks([0.001, 0.01, 0.1, 1], ['0.001', '0.01', '0.1', '1'])
plt.ylabel('Tiempo (segundos)', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV con los resultados de Speedup y Eficiencia
speedup_eficiencia_df = pd.read_csv('../pruebas/speedup_eficiencia_promedios_cuda.csv')

# Gráfico de Speedup vs Num-Bloques
plt.figure(figsize=(10, 6))
plt.plot(speedup_eficiencia_df['Bloques'], speedup_eficiencia_df['Speedup'], marker='s', linestyle='-', color='g', label='Speedup')
plt.title('Speedup vs Número de Bloques', fontsize=14)
plt.xlabel('Número de Bloques', fontsize=12)
plt.ylabel('Speedup', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()

# Gráfico de Eficiencia vs Num-Bloques
plt.figure(figsize=(10, 6))
plt.plot(speedup_eficiencia_df['Bloques'], speedup_eficiencia_df['Eficiencia'], marker='^', linestyle='-', color='r', label='Eficiencia')
plt.title('Eficiencia vs Número de Bloques', fontsize=14)
plt.xlabel('Número de Bloques', fontsize=12)
plt.ylabel('Eficiencia', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV sin encabezado
tiempos_df = pd.read_csv('../pruebas/tiempos_vs_n_openmp.csv', header=None)

# Asignar nombres de columnas manualmente
tiempos_df.columns = ['n', 'tiempo']

# Calcular el promedio de tiempo para cada valor único de n
promedios_df = tiempos_df.groupby('n', as_index=False).mean()

# Verificar las primeras filas del dataframe promedio
print(promedios_df.head(10))

# Gráfico de Tiempo vs n
plt.figure(figsize=(10, 6))
plt.plot(promedios_df['n'], promedios_df['tiempo'], marker='o', linestyle='-', color='b', label='Promedio Tiempo de ejecución')
plt.title('Tiempo Promedio vs Tamaño de Entrada (n)', fontsize=14)
plt.xlabel('Tamaño de Entrada (n)', fontsize=12)
plt.xscale('log')  # Escala logarítmica en el eje X
plt.yscale('log')
plt.xticks([100, 1000, 10000, 100000, 1000000, 10000000], ['100', '1k', '10k', '100k', '1M', '10M'])
plt.yticks([0.1, 1, 10], ['0.1', '1', '10'])
plt.ylabel('Tiempo Promedio (segundos)', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
